```python
from pandas import Series, DataFrame
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
e1 = pd.read_table("/Volumes/Inugoya0/shizuokangs_vol1/kallisto_test_result/ERR1551404/abundance.tsv")
e1 = e1.drop(['length', 'eff_length', 'est_counts'], axis=1)
e1.columns = ['target_id', 'TPM_ERR1551404']

e2 = pd.read_table("/Volumes/Inugoya0/shizuokangs_vol1/kallisto_test_result/ERR1551408/abundance.tsv")
e2 = e2.drop(['length', 'eff_length', 'est_counts'], axis=1)
e2.columns = ['target_id', 'TPM_ERR1551408']

e = pd.merge(e1, e2, on='target_id')
# トランスクリプトのidの小数点以下を除いておく。
e['target_id'] = e['target_id'].apply(lambda x: x.split('.')[0])

e['log_ERR1551404'] = np.log10(e['TPM_ERR1551404'] + 1)
e['log_ERR1551408'] = np.log10(e['TPM_ERR1551408'] + 1)
e['diff'] = abs(e['log_ERR1551404'] - e['log_ERR1551408'])

e.head()
```

```python
e.hist(column='diff')
plt.show()
```
